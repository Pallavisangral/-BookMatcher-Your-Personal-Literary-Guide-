```python
import spacy
import pandas as pd
```

```python
import numpy as np
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
data = pd.read_csv("/content/BookandDesc (2).csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
text = data['summary'].tolist()

text
```

```python
noun_chunks = []

for i in text:
    doc = nlp(i)
    noun_chunks.append(" ".join([chunk.text for chunk in doc.noun_chunks]))
```

add in dataframe


```python
data['keyphrases'] = noun_chunks
```

```python
data.head(5)
```

Data preprocessing

```python
data.keyphrases = data.keyphrases.apply(lambda x: x.lower())
```

```python
import string
string.punctuation
#defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
#storing the puntuation free text
data.keyphrases = data.keyphrases.apply(lambda x:remove_punctuation(x))
```

```python
data.head(5)
```

Tokenisation

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
import pandas as pd



# Define a tokenization function
def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens

# Apply tokenization on the 'text' column
data['tokenized_text'] = data['keyphrases'].apply(tokenize_text)


```

```python
data.head(5)
```

```python
data.head(5)
```

```python
data['tokenized_text'] = data['tokenized_text'].astype(str)
```

```python
data.head(5)
```

word embedding

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(data['tokenized_text'])
features = vectorizer.get_feature_names_out()
print(features)

```

```python
len(features)
```

```python
tfidf_array = X.toarray()
```

```python
from sklearn.metrics.pairwise import  cosine_similarity
```

```python
cosine_sim = cosine_similarity(tfidf_array)
```

```python
import pickle
# Open a file and use dump()
with open('similarityfile.pkl', 'wb') as file:
      
    # A new file will be created
    pickle.dump(cosine_sim, file)
```

```python
def recommend(book):
    # idx = indices[book]
    index = data[data['isbn'] == book].index[0]
    distances = sorted(list(enumerate(cosine_sim[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(i, data.iloc[i[0]].title)
    
        
```

```python
recommend('679810307')
```
