```python
completedata0 = pd.merge(books, description , on = 'ISBN', how = 'inner')
```

```python
completedata0
```

```python
completedata = pd.merge(completedata0,rating, on= 'ISBN', how ='inner')
```

```python
completedata
```

```python
rating_mod=completedata.groupby("User-ID").count()['Book-Rating'].reset_index(name="count")
rating_mod
```

```python
rating_mod_ = rating_mod[rating_mod['count'] >=100]
```

```python
rating_mod_
```

```python
rating_mod_merge = pd.merge(rating,rating_mod_, on = 'User-ID', how = 'inner')
```

```python
rating_mod_merge
```

```python
rating_modified = rating_mod_merge.groupby('ISBN').count()['User-ID'].reset_index(name ='mostratedbooks')
```

```python
rating_modified
```

```python
rating_modified_ = rating_modified[rating_modified['mostratedbooks']>=40]
```

```python
rating_modified
```

```python
final_data_mod =rating_mod_merge.drop(['count'], axis =1)
```

```python
final_data_mod
```

```python
remdup = final_data_mod.drop_duplicates(subset = ['User-ID', 'ISBN'])
```

```python
remdup
```

```python
final_data_mod.duplicated().sum()
```

```python
# final_book_rating= final_data_mod.set_index('User-ID')
```

```python
final_book_rating = pd.pivot_table(remdup, values= 'Book-Rating', index = ['ISBN'], columns = ['User-ID'])
```

```python
final_book_rating
```

```python
final_book_rating.fillna(0, inplace = True)
```
