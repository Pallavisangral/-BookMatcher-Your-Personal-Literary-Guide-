count of number of users rating the book

```python
counted=completebook.groupby("ISBN").count()['Book-Rating'].reset_index(name="rating-count")
counted
```

```python
# completebook1=pd.merge(counted,completebook, on='ISBN', how= "inner")
```

count averge of complete book

```python
average=completebook.groupby("ISBN").mean()['Book-Rating'].reset_index(name="average-rating")
average
```

```python
merged = pd.merge(counted,average, on = 'ISBN', how = 'inner')
merged.head(5)
```
