import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

get data

```python
df = pd.read_csv("correctdf.csv" )
```

```python
df.head(5)
```

```python
df.shape
```

```python
df1 = pd.pivot_table(df, values= 'Book-Rating', index = ['ISBN'], columns = ['User-ID'])
```

```python
df1
```

```python
df1.fillna(0, inplace = True)
```

```python
df1
```

```python
df1.shape
```

```python
from scipy.sparse import csr_matrix
```

```python
df2matrix = csr_matrix(df1.values)

```

```python
print(df2matrix)
```

```python
from sklearn.neighbors import NearestNeighbors


model = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model.fit(df2matrix)
```

```python
df2matrix.shape
```

```python
randomvalue = np.random.choice(df1.shape[0])
print(randomvalue)
```

```python
gap, indices = model.kneighbors(df1.iloc[randomvalue,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
print(gap)
```

```python
for i in range(0, len(gap.flatten())):
    if i == 0:
        print('Recommendations for ISBN {0}:\n'.format(df1.index[randomvalue]))
    else:
        print('ISBN {0}: {1}, with distance of {2}:'.format(i, df1.index[indices.flatten()[i]], gap.flatten()[i]))
```
